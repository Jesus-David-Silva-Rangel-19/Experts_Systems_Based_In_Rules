# **Riesgo Probabilistico De Tener Depresión**

```python
# Definir una función para estimar el riesgo de depresión
def estimar_riesgo_depresion(edad, genero, historial_familiar, sintomas):
    riesgo = 0
    
    # Regla 1: Edad
    if edad >= 45:
        riesgo += 20
    elif edad >= 30:
        riesgo += 10
    else:
        riesgo += 5
    
    # Regla 2: Género
    if genero == 'F':
        riesgo += 10
    
    # Regla 3: Historial familiar
    if historial_familiar:
        riesgo += 30
    
    # Regla 4: Síntomas
    if 'tristeza' in sintomas:
        riesgo += 20
    if 'insomnio' in sintomas:
        riesgo += 15
    if 'fatiga' in sintomas:
        riesgo += 10
    if 'falta de concentración' in sintomas:
        riesgo += 10

    # Regla 5: Demografía
    if edad >= 45 and genero == 'F':
        riesgo += 10
    if edad >= 45 and genero == 'M':
        riesgo += 5
    if edad >= 30 and genero == 'F':
        riesgo += 5
    if edad >= 30 and genero == 'M':
        riesgo += 2.5

    # Regla 6: Síntomas específicos
    if 'pérdida de apetito' in sintomas:
        riesgo += 10
    if 'pérdida de peso' in sintomas:
        riesgo += 10
    if 'pensamientos suicidas' in sintomas:
        riesgo += 20
    if 'ansiedad' in sintomas:
        riesgo += 15
    if 'irritabilidad' in sintomas:
        riesgo += 10
    if 'dolores corporales' in sintomas:
        riesgo += 10
    
    # Calcular el riesgo total
    if riesgo > 100:
        riesgo = 100
    
    return riesgo

# Ejemplo de uso
edad = 21
genero = 'M'
historial_familiar = True
sintomas = ['tristeza', 'insomnio', 'fatiga', 'falta de concentración', 'pérdida de apetito', 'pérdida de peso', 'pensamientos suicidas', 'ansiedad', 'irritabilidad', 'dolores corporales']

riesgo = estimar_riesgo_depresion(edad, genero, historial_familiar, sintomas)
print(f'El riesgo estimado de tener depresión es: {riesgo}%')
```

```python
def diagnosticar_diabetes(edad, genero, historial_familiar, sintomas, factores_riesgo, hba1c, glucosa_ayunas):
    tipo_diabetes = None
    probabilidad = 0
    
    # Regla 1: HbA1c - Indicador fuerte de diabetes
    if hba1c >= 6.5:
        probabilidad += 45
    elif hba1c >= 5.7:
        probabilidad += 25
    
    # Regla 2: Glucosa en ayunas - Indicador fuerte de diabetes
    if glucosa_ayunas >= 126:
        probabilidad += 40
    elif glucosa_ayunas >= 100:
        probabilidad += 20
    
    # Regla 3: Edad - Factor que aumenta el riesgo
    if edad >= 60:
        probabilidad += 15
    elif edad >= 45:
        probabilidad += 10
    elif edad >= 30:
        probabilidad += 5

    # Regla 4: Género - Riesgo levemente mayor para mujeres en algunos estudios
    if genero == 'F':
        probabilidad += 5
    
    # Regla 5: Historial familiar - Importante para el riesgo general
    if historial_familiar:
        probabilidad += 20
    
    # Regla 6: Factores de riesgo - Obesidad tiene un impacto significativo
    if 'obesidad' in factores_riesgo:
        probabilidad += 25
    if 'hipertension' in factores_riesgo:
        probabilidad += 10
    if 'sedentarismo' in factores_riesgo:
        probabilidad += 10
    
    # Regla 7: Síntomas - Indicadores comunes de diabetes
    if 'aumento de la sed' in sintomas:
        probabilidad += 15
    if 'aumento de la frecuencia urinaria' in sintomas:
        probabilidad += 15
    if 'fatiga' in sintomas:
        probabilidad += 10
    if 'visión borrosa' in sintomas:
        probabilidad += 10
    
    # Diagnóstico basado en la probabilidad
    if probabilidad >= 90:
        tipo_diabetes = 'Diabetes Tipo 1'
    elif probabilidad >= 70:
        tipo_diabetes = 'Diabetes Tipo 2'
    elif probabilidad >= 50:
        tipo_diabetes = 'Prediabetes'
    else:
        tipo_diabetes = 'No diabetes'

    # Limitar la probabilidad total a 100
    probabilidad = min(probabilidad, 100)
    
    return tipo_diabetes, probabilidad

# Ejemplo de uso
hba1c = 3.8
glucosa_ayunas = 10
factores_riesgo = ['sedentarismo']
sintomas = []
edad = 21
genero = 'M'
historial_familiar = True

tipo_diabetes, probabilidad_diabetes = diagnosticar_diabetes(edad, genero, historial_familiar, sintomas, factores_riesgo, hba1c, glucosa_ayunas)

print(f'El diagnóstico es: {tipo_diabetes} con una probabilidad del {probabilidad_diabetes}%')
```

# **Enfermedad Coronoria (EC)**

```python
def estimar_riesgo_enfermedad_coronaria(edad, genero, hipertension, colesterol, diabetes, tabaquismo, obesidad, sedentarismo, historial_familiar, sintomas):
    # Inicialización del puntaje de riesgo
    riesgo_ec = 0
    
    # Regla 1: Edad - Aumenta el riesgo a partir de los 45 años en hombres y 55 en mujeres
    if (genero == 'M' and edad >= 45) or (genero == 'F' and edad >= 55):
        riesgo_ec += 15
    
    # Regla 2: Hipertensión - Factor de riesgo importante para la EC
    if hipertension:
        riesgo_ec += 20
    
    # Regla 3: Colesterol elevado - LDL alto (>130 mg/dL) es un factor de riesgo importante
    if colesterol == 'alto':
        riesgo_ec += 20
    elif colesterol == 'moderado':
        riesgo_ec += 10
    
    # Regla 4: Diabetes - Incrementa significativamente el riesgo cardiovascular
    if diabetes:
        riesgo_ec += 20
    
    # Regla 5: Tabaquismo - Factor de riesgo conocido para enfermedad coronaria
    if tabaquismo:
        riesgo_ec += 15
    
    # Regla 6: Obesidad - Indicador significativo cuando el IMC es ≥30
    if obesidad:
        riesgo_ec += 10
    
    # Regla 7: Sedentarismo - Contribuye a varios factores de riesgo
    if sedentarismo:
        riesgo_ec += 10
    
    # Regla 8: Historial familiar - Aumenta el riesgo en personas con antecedentes de EC en parientes cercanos
    if historial_familiar:
        riesgo_ec += 10
    
    # Regla 9: Síntomas - Indicativos de posible EC
    if 'dolor torácico' in sintomas:
        riesgo_ec += 20
    if 'disnea' in sintomas:
        riesgo_ec += 15
    if 'fatiga' in sintomas:
        riesgo_ec += 10
    if 'palpitaciones' in sintomas:
        riesgo_ec += 5
    
    # Determinar diagnóstico textual basado en el riesgo acumulado
    if riesgo_ec >= 90:
        diagnostico = 'Alta probabilidad de enfermedad coronaria'
        probabilidad = 95  # Probabilidad aproximada de tener EC
    elif riesgo_ec >= 70:
        diagnostico = 'Probabilidad moderada de enfermedad coronaria'
        probabilidad = 75
    elif riesgo_ec >= 40:
        diagnostico = 'Probabilidad baja de enfermedad coronaria'
        probabilidad = 50
    else:
        diagnostico = 'Riesgo bajo de enfermedad coronaria'
        probabilidad = 20
    
    # Limitar el riesgo acumulado y probabilidad máxima a 100
    riesgo_ec = min(riesgo_ec, 100)
    probabilidad = min(probabilidad, 100)
    
    return diagnostico, riesgo_ec, probabilidad

# Ejemplo de uso
edad = 21
genero = 'M'
hipertension = False
colesterol = 'moderado'
diabetes = False
tabaquismo = False
obesidad = False
sedentarismo = True
historial_familiar = False
sintomas = []

diagnostico_ec, riesgo_ec, probabilidad_ec = estimar_riesgo_enfermedad_coronaria(edad, genero, hipertension, colesterol, diabetes, tabaquismo, obesidad, sedentarismo, historial_familiar, sintomas)

print(f'Diagnóstico: {diagnostico_ec} con un riesgo estimado de {riesgo_ec}% y una probabilidad de enfermedad coronaria de {probabilidad_ec}%')
```

# **Trastorno Obsesivo-Compulsivo**

```python
def evaluar_toc(edad, genero, antecedentes_familiares, factores_riesgo, sintomas, lista_verificacion_dsm5):
    # Inicializar variables de resultado
    diagnostico_toc = None
    tipo_obsesion = None
    tipo_compulsion = None
    riesgo_toc = 0

    # Factores de riesgo adicionales
    if edad <= 25:  # Mayor riesgo en adultos jóvenes
        riesgo_toc += 10
    if genero == 'F':
        riesgo_toc += 5  # Las mujeres pueden experimentar TOC más frecuentemente en ciertos tipos de obsesiones
    if antecedentes_familiares:
        riesgo_toc += 15  # Factor hereditario significativo
    
    # Evaluación de factores de riesgo específicos
    if 'historial de ansiedad' in factores_riesgo:
        riesgo_toc += 15
    if 'historial de depresión' in factores_riesgo:
        riesgo_toc += 10
    if 'eventos traumáticos' in factores_riesgo:
        riesgo_toc += 10
    if 'estrés crónico' in factores_riesgo:
        riesgo_toc += 10
    
    # Evaluación de síntomas de obsesión
    if 'miedo a la contaminación' in sintomas:
        riesgo_toc += 15
        tipo_obsesion = 'Obsesión de contaminación'
    elif 'duda y necesidad de comprobación' in sintomas:
        riesgo_toc += 15
        tipo_obsesion = 'Obsesión de comprobación'
    elif 'pensamientos intrusivos' in sintomas:
        riesgo_toc += 15
        tipo_obsesion = 'Obsesión de pensamientos intrusivos'
    elif 'preocupación por el orden' in sintomas:
        riesgo_toc += 15
        tipo_obsesion = 'Obsesión de orden y simetría'
    
    # Evaluación de síntomas de compulsión
    if 'lavado de manos frecuente' in sintomas:
        riesgo_toc += 15
        tipo_compulsion = 'Compulsión de limpieza'
    elif 'comprobaciones reiteradas' in sintomas:
        riesgo_toc += 15
        tipo_compulsion = 'Compulsión de comprobación'
    elif 'repetición de acciones' in sintomas:
        riesgo_toc += 15
        tipo_compulsion = 'Compulsión de repetición'
    elif 'organización excesiva' in sintomas:
        riesgo_toc += 15
        tipo_compulsion = 'Compulsión de orden'
    
    # Verificación según el DSM-5
    if 'obsesiones intrusivas' in lista_verificacion_dsm5:
        riesgo_toc += 10
    if 'compulsiones difíciles de controlar' in lista_verificacion_dsm5:
        riesgo_toc += 10
    if 'afectación de la vida diaria' in lista_verificacion_dsm5:
        riesgo_toc += 10
    if 'comportamientos repetitivos o ritualizados' in lista_verificacion_dsm5:
        riesgo_toc += 10

    # Diagnóstico basado en el riesgo total acumulado
    if riesgo_toc >= 75:
        diagnostico_toc = 'Posible Trastorno Obsesivo-Compulsivo'
        probabilidad_toc = 90  # Probabilidad aproximada de TOC
    elif riesgo_toc >= 50:
        diagnostico_toc = 'Probabilidad moderada de TOC'
        probabilidad_toc = 70
    elif riesgo_toc >= 30:
        diagnostico_toc = 'Baja probabilidad de TOC'
        probabilidad_toc = 40
    else:
        diagnostico_toc = 'Riesgo bajo de TOC'
        probabilidad_toc = 20
    
    # Limitar el riesgo acumulado a un máximo de 100
    riesgo_toc = min(riesgo_toc, 100)
    probabilidad_toc = min(probabilidad_toc, 100)
    
    return diagnostico_toc, tipo_obsesion, tipo_compulsion, riesgo_toc, probabilidad_toc

# Ejemplo de uso
edad = 21
genero = 'M'
antecedentes_familiares = False
factores_riesgo = ['historial de ansiedad', 'estrés crónico']
sintomas = ['pensamientos intrusivos', 'repetición de acciones']
lista_verificacion_dsm5 = ['obsesiones intrusivas', 'compulsiones difíciles de controlar', 'afectación de la vida diaria']

diagnostico, tipo_obsesion, tipo_compulsion, riesgo_toc, probabilidad_toc = evaluar_toc(edad, genero, antecedentes_familiares, factores_riesgo, sintomas, lista_verificacion_dsm5)

print(f'Diagnóstico: {diagnostico}')
print(f'Tipo de obsesión: {tipo_obsesion}')
print(f'Tipo de compulsión: {tipo_compulsion}')
print(f'Riesgo estimado de TOC: {riesgo_toc}%')
print(f'Probabilidad de TOC: {probabilidad_toc}%')
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir parámetros para el modelo SEIR
def modelo_seir(y, t, beta, sigma, gamma, N):
    S, E, I, R = y
    dS_dt = -beta * S * I / N                # Tasa de cambio de susceptibles
    dE_dt = beta * S * I / N - sigma * E     # Tasa de cambio de expuestos
    dI_dt = sigma * E - gamma * I            # Tasa de cambio de infectados
    dR_dt = gamma * I                        # Tasa de cambio de recuperados
    return [dS_dt, dE_dt, dI_dt, dR_dt]

# Definir la función para simular la propagación de la enfermedad
def simular_monkeypox(poblacion, infectados_iniciales, expuestos_iniciales, recuperados_iniciales,
                      beta, sigma, gamma, dias):
    # Condiciones iniciales
    susceptibles_iniciales = poblacion - infectados_iniciales - expuestos_iniciales - recuperados_iniciales
    y0 = [susceptibles_iniciales, expuestos_iniciales, infectados_iniciales, recuperados_iniciales]
    
    # Definir el vector de tiempo
    t = np.linspace(0, dias, dias)
    
    # Resolver el sistema de ecuaciones diferenciales
    solucion = odeint(modelo_seir, y0, t, args=(beta, sigma, gamma, poblacion))
    S, E, I, R = solucion.T
    
    # Graficar resultados
    plt.figure(figsize=(6, 4))
    plt.plot(t, S, label='Susceptibles', color='blue')
    plt.plot(t, E, label='Expuestos', color='orange')
    plt.plot(t, I, label='Infectados', color='red')
    plt.plot(t, R, label='Recuperados', color='green')
    plt.xlabel('Días')
    plt.ylabel('Número de personas')
    plt.title('Simulación de propagación de Monkeypox con modelo SEIR')
    plt.legend(loc='best')
    plt.grid()
    plt.show()

    return t, S, E, I, R

# Parámetros para la simulación
poblacion = 1000000             # Tamaño de la población
infectados_iniciales = 1        # Personas inicialmente infectadas
expuestos_iniciales = 0         # Personas inicialmente expuestas
recuperados_iniciales = 0       # Personas inicialmente recuperadas

beta = 0.3                      # Tasa de transmisión
sigma = 1/7                     # Tasa de progresión de expuestos a infectados (1/periodo de incubación)
gamma = 1/14                    # Tasa de recuperación (1/periodo infeccioso)

dias = 160                      # Número de días a simular

# Ejecutar la simulación
t, S, E, I, R = simular_monkeypox(poblacion, infectados_iniciales, expuestos_iniciales,
                                  recuperados_iniciales, beta, sigma, gamma, dias)
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Definir el modelo SEIR
def modelo_seir(y, t, beta, sigma, gamma, N):
    S, E, I, R = y
    dS_dt = -beta * S * I / N                
    dE_dt = beta * S * I / N - sigma * E     
    dI_dt = sigma * E - gamma * I            
    dR_dt = gamma * I                        
    return [dS_dt, dE_dt, dI_dt, dR_dt]

# Configurar la función de simulación con tiempo en horas
def simular_monkeypox(poblacion, infectados_iniciales, expuestos_iniciales, recuperados_iniciales,
                      beta, sigma, gamma, horas):
    susceptibles_iniciales = poblacion - infectados_iniciales - expuestos_iniciales - recuperados_iniciales
    y0 = [susceptibles_iniciales, expuestos_iniciales, infectados_iniciales, recuperados_iniciales]
    
    t = np.linspace(0, horas, 200)  # 200 puntos en 20 segundos
    
    solucion = odeint(modelo_seir, y0, t, args=(beta, sigma, gamma, poblacion))
    S, E, I, R = solucion.T
    
    return t, S, E, I, R

# Parámetros para la simulación
poblacion = 17504
infectados_iniciales = 1
expuestos_iniciales = 0
recuperados_iniciales = 0

beta = 0.3
sigma = 1/168  # Tasa de incubación en horas
gamma = 1/336  # Tasa de recuperación en horas

horas = 160 * 24  # Simulación de hasta 160 días en horas

# Ejecutar la simulación
t, S, E, I, R = simular_monkeypox(poblacion, infectados_iniciales, expuestos_iniciales,
                                  recuperados_iniciales, beta, sigma, gamma, horas)

# Configurar la animación
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xlim(0, horas)
ax.set_ylim(0, poblacion)
ax.set_xlabel('Tiempo (horas)')
ax.set_ylabel('Número de personas')
ax.set_title('Propagación de Monkeypox (Modelo SEIR)')

line_S, = ax.plot([], [], label='Susceptibles', color='blue')
line_E, = ax.plot([], [], label='Expuestos', color='orange')
line_I, = ax.plot([], [], label='Infectados', color='red')
line_R, = ax.plot([], [], label='Recuperados', color='green')

plt.legend()

# Función de actualización para la animación
def actualizar(frame):
    line_S.set_data(t[:frame], S[:frame])
    line_E.set_data(t[:frame], E[:frame])
    line_I.set_data(t[:frame], I[:frame])
    line_R.set_data(t[:frame], R[:frame])
    return line_S, line_E, line_I, line_R

# Crear la animación, durando exactamente 20 segundos
anim = FuncAnimation(fig, actualizar, frames=100, interval=100, blit=True)

# Mostrar la animación en un Jupyter Notebook
plt.show()

# Si deseas guardar la animación
anim.save("propagacion_monkeypox_jdsr.gif", writer="imagemagick")
# anim.save("propagacion_monkeypox.mp4", writer="ffmpeg")
```
